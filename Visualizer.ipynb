```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
 
df = None
 
while True:
    print("\n========== Data Analysis & Visualization Program ==========")
    print("1. Load Dataset")
    print("2. Explore Data")
    print("3. Perform DataFrame Operation")
    print("4. Handle Missing Data")
    print("5. Generate Descriptive Statistics")
    print("6. Data Visualization")
    print("7. Save Visualization")
    print("8. Exit")
 
    try:
        c = int(input("\nEnter Your Choice -> "))
    except:
        print("Please enter a valid number.")
        continue
 
    # -------------------- Load Dataset --------------------
    if c == 1:
 
        try:
            path = input("Enter CSV File Path : ")
            df = pd.read_csv(path)
            print("\nDataset Loaded Successfully.")
            print("Rows :", df.shape[0])
            print("Columns :", df.shape[1])
 
        except Exception as e:
            print("Error :", e)
 
    # -------------------- Explore Data --------------------
 
    elif c == 2:
 
        if df is None:
            print("Please Load Dataset First.")
            continue
 
        while True:
 
            print("\n========== Explore Data ==========")
            print("1. Display First 5 Rows")
            print("2. Display Last 5 Rows")
            print("3. Display Column Names")
            print("4. Display Data Types")
            print("5. Display Basic Information")
            print("6. Back")
 
            try:
                i = int(input("Enter Choice : "))
            except:
                print("Invalid Input")
                continue
 
            if i == 1:
                print(df.head())
 
            elif i == 2:
                print(df.tail())
 
            elif i == 3:
                print("\nColumn Names\n")
                for col in df.columns:
                    print(col)
 
            elif i == 4:
                print(df.dtypes)
 
            elif i == 5:
                print(df.info())
 
            elif i == 6:
                break
 
            else:
                print("Invalid Choice")
 
    # -------------------- DataFrame Operation --------------------
 
    elif c == 3:
 
        if df is None:
            print("Please Load Dataset First.")
            continue
 
        while True:
 
            print("\n========== DataFrame Operation ==========")
            print("1. Check Missing Values")
            print("2. Describe Dataset")
            print("3. Back")
 
            try:
                m = int(input("Enter Choice : "))
            except:
                print("Invalid Input")
                continue
 
            if m == 1:
                print(df.isnull().sum())
 
            elif m == 2:
                print(df.describe(include="all"))
 
            elif m == 3:
                break
 
            else:
                print("Invalid Choice")
 
    # -------------------- Handle Missing Data --------------------
 
    elif c == 4:
 
        if df is None:
            print("Please Load Dataset First.")
            continue
 
        while True:
 
            print("\n========== Handle Missing Data ==========")
            print("1. Display Rows with Missing Values")
            print("2. Fill Missing Values with Mean")
            print("3. Drop Rows with Missing Values")
            print("4. Replace Missing Values with Specific Value")
            print("5. Back")
 
            try:
                h = int(input("Enter Choice : "))
            except:
                print("Invalid Input")
                continue
 
            if h == 1:
 
                missing = df[df.isnull().any(axis=1)]
 
                if missing.empty:
                    print("No Missing Values Found.")
                else:
                    print(missing)
 
            elif h == 2:
 
                numeric_columns = df.select_dtypes(include=np.number).columns
 
                df[numeric_columns] = df[numeric_columns].fillna(
                    df[numeric_columns].mean()
                )
 
                print("\nMissing Numeric Values Filled Successfully.")
                print(df.head())
 
            elif h == 3:
 
                df = df.dropna()
 
                print("Rows with Missing Values Removed Successfully.")
                print("Current Shape :", df.shape)
 
            elif h == 4:
 
                value = input("Enter Replacement Value : ")
 
                df = df.fillna(value)
 
                print("Missing Values Replaced Successfully.")
 
            elif h == 5:
                break
 
            else:
                print("Invalid Choice")
 
    # -------------------- Descriptive Statistics --------------------
 
    elif c == 5:
 
        if df is None:
            print("Please Load Dataset First.")
            continue
 
        print("\n========== Descriptive Statistics ==========\n")
        print(df.describe(include="all"))
 
    # -------------------- Data Visualization --------------------
 
    elif c == 6:
 
        if df is None:
            print("Please Load Dataset First.")
            continue
 
        while True:
 
            print("\n========== Data Visualization ==========")
            print("1. Bar Plot")
            print("2. Line Plot")
            print("3. Scatter Plot")
            print("4. Pie Chart")
            print("5. Histogram")
            print("6. Stack Plot")
            print("7. Back")
 
            try:
                v = int(input("Enter Your Choice : "))
            except:
                print("Invalid Input")
                continue
 
            if v == 7:
                break
 
            if v not in [1, 2, 3, 4, 5, 6]:
                print("Invalid Choice")
                continue
 
            x = input("Enter X-axis Column Name : ")
 
            if x not in df.columns:
                print("Invalid Column Name")
                continue
 
            if v != 5:
                y = input("Enter Y-axis Column Name : ")
 
                if y not in df.columns:
                    print("Invalid Column Name")
                    continue
 
            plt.figure(figsize=(10, 6))
 
            try:
 
                # ---------------- Bar Plot ----------------
 
                if v == 1:
 
                    plt.bar(df[x], df[y], color="skyblue")
                    plt.xlabel(x)
                    plt.ylabel(y)
                    plt.title("Bar Plot")
 
                # ---------------- Line Plot ----------------
 
                elif v == 2:
 
                    plt.plot(df[x], df[y], marker="o", color="green")
                    plt.xlabel(x)
                    plt.ylabel(y)
                    plt.title("Line Plot")
 
                # ---------------- Scatter Plot ----------------
 
                elif v == 3:
 
                    plt.scatter(df[x], df[y], color="red")
                    plt.xlabel(x)
                    plt.ylabel(y)
                    plt.title("Scatter Plot")
 
                # ---------------- Pie Chart ----------------
 
                elif v == 4:
 
                    plt.pie(df[y],
                            labels=df[x],
                            autopct="%1.1f%%",
                            startangle=90)
 
                    plt.title("Pie Chart")
 
                # ---------------- Histogram ----------------
 
                elif v == 5:
 
                    sns.histplot(df[x], bins=20, kde=True)
                    plt.xlabel(x)
                    plt.title("Histogram")
 
                # ---------------- Stack Plot ----------------
 
                elif v == 6:
 
                    plt.stackplot(range(len(df)),
                                  df[y],
                                  labels=[y])
 
                    plt.legend(loc="upper left")
                    plt.title("Stack Plot")
                    plt.xlabel("Index")
                    plt.ylabel(y)
 
                plt.tight_layout()
                plt.show()
 
            except Exception as e:
                print("Error :", e)
 
    # -------------------- Save Visualization --------------------
 
    elif c == 7:
 
        if plt.get_fignums():
 
            filename = input("Enter File Name (Example : graph.png) : ")
            plt.savefig(filename)
            print("Graph Saved Successfully as", filename)
 
        else:
            print("Please Generate a Graph First.")
 
    # -------------------- Exit --------------------
 
    elif c == 8:
 
        print("\nThank You For Using Data Analysis & Visualization Program.")
        print("Program Closed Successfully.")
        break
 
    else:
        print("Invalid Choice. Please Try Again.")
 
```
